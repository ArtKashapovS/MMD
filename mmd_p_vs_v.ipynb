```python
import matplotlib.pyplot as plt
```

```python
vs = [4.206649103322365, 5.0479789239868374, 5.88930874465131, 6.730638565315785, 7.571968385980257, 8.41329820664473]
vols = [v * 10000 for v in vs]
```

```python
ps = [25777891.057390798, 16879318.35620919, 14961971.630774466, 12927272.159270016, 11246701.908118377, 9840532.758256653]
pres = [p / 1000000 for p in ps]
```

```python
plt.figure(figsize=(16, 8))
plt.plot(vols, pres, 'ro', lw=15)
plt.xlabel('Объем, au')
plt.ylabel('Давление, au')
# plt.xlim([vols[0], vols[-1]])
plt.grid()
plt.show()
```
